### For the first dashboard

```python
# We want to find the sum of order payments for each year. We will join two tables - order_status_year_price and
# order_payments.

query = "SELECT * FROM olist.order_status_year_price Inner Join olist.order_payments on order_status_year_price.order_id = order_payments.order_id "
order_year_price_join = pd.read_sql_query(query, db_connection)
order_year_price_join
```

```python
# Cleaning the column

# I will delete the columns we don't need.

order_year_price_join.drop(columns = ["category", "order_item_id", "order_status", "payment_sequential", "payment_type", "payment_installments"], inplace = True)

order_year_price_join
```

```python
# Calculating the nr of sellers.

query_nr_sellers = "SELECT count(seller_id) FROM olist.sellers"
nr_of_sellers = pd.read_sql_query(query_nr_sellers, db_connection)
nr_of_sellers
```

```python
# Calculating the nr of customers.

query_nr_customers = "SELECT count(customer_id) FROM olist.customers"
nr_of_customers = pd.read_sql_query(query_nr_customers, db_connection)
nr_of_customers
```

```python
# Calculating the avg sales amount per customer

avg_sales_per_customer = (order_year_price_join["payment_value"].sum()) / nr_of_customers
avg_sales_per_customer
```

```python
# Top categories by sales amount

query_top_categories = "SELECT * FROM olist.order_payments Inner join olist.order_items on order_payments.order_id = order_items.order_id"
top_categories_first = pd.read_sql_query(query_top_categories, db_connection)
top_categories_first
```

```python
query_top_categories_new = "SELECT * FROM olist.products Inner join olist.product_category_name_translation on products.product_category_name = product_category_name_translation.product_category_name"
top_categories_second = pd.read_sql_query(query_top_categories_new, db_connection)
top_categories_second
```

```python
top_categories_df = pd.merge(top_categories_first, top_categories_second, on = "product_id")
top_categories_df
```

```python
top_categories_df.drop(columns = ["payment_sequential", "payment_type", "payment_installments", "order_item_id", "shipping_limit_date", "product_name_length", "product_description_length", "product_photos_qty", "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"], inplace = True)

top_categories_df
```

```python
top_categories_df.isna().sum()
```

```python
top_categories_df.describe()
```

```python
# Top 10 categories by sales amount.

category_sold = top_categories_df.groupby(["product_category_name_english"], as_index=False)["payment_value"].count().rename(columns = {"payment_value":"total_value"})
category_sold["%"] = round((category_sold["total_value"] / category_sold["total_value"].sum()) * 100, 2)
category_sold10 = category_sold.nlargest(10, "total_value")
category_sold10
```

```python
# Top 10 sellers.

query_for_top_sellers = "SELECT * FROM olist.order_items inner join olist.order_payments on order_items.order_id = order_payments.order_id"
top_sellers_first = pd.read_sql_query(query_for_top_sellers, db_connection)
top_sellers_first
```

```python
query_for_top_sellers_two = "SELECT seller_id FROM olist.sellers"
top_sellers_second = pd.read_sql_query(query_for_top_sellers_two, db_connection)
top_sellers_second
```

```python
top_sellers_df = pd.merge(top_sellers_first, top_sellers_second, on = "seller_id")
top_sellers_df
```

```python
top_sellers_df.drop(columns = ["order_item_id", "shipping_limit_date", "payment_sequential", "payment_type", "payment_installments"], inplace = True)

top_sellers_df
```
