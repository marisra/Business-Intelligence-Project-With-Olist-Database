#### Here I will start joining, exploring and cleaning all tables which I need for for the first dashboard.

```python
# We want to find the sum of order payments for each year. We will join two tables - order_status_year_price and
# order_payments. I will use sql queries to do that.

query = "SELECT * FROM olist.order_status_year_price Inner Join olist.order_payments on order_status_year_price.order_id = order_payments.order_id "
order_year_price_join = pd.read_sql_query(query, db_connection)
order_year_price_join
```

```python
# I will delete the columns we don't need.

order_year_price_join.drop(columns = ["category", "order_item_id", "order_status", "payment_sequential", "payment_type", "payment_installments"], inplace = True)

order_year_price_join
```

```python
# Checking missing values

order_year_price_join.isna().sum()
```

```python
# Checking the data types

order_year_price_join.dtypes
```

```python
# Here I am checking the year column to see which years do we have and if there is anything to standardize.

order_year_price_join["Year"].unique()
```

```python
# Overview of some of the descriptive statistics in this table.

order_year_price_join.describe()
```

```python
# I will calculate the total sales for each year

order_year_price_join.groupby("Year").sum()
```

```python
# Calculating the unique nr of sellers for all the years. I will use another table.

query_nr_sellers = "SELECT count(distinct seller_id) FROM olist.sellers"
nr_of_sellers = pd.read_sql_query(query_nr_sellers, db_connection)
nr_of_sellers
```

```python
# Calculating the unique nr of customers for all the years.

query_nr_customers = "SELECT count(distinct customer_id) FROM olist.customers"
nr_of_customers = pd.read_sql_query(query_nr_customers, db_connection)
nr_of_customers
```

```python
# Calculating the avg sales amount per customer.

avg_sales_per_customer = (order_year_price_join["payment_value"].sum()) / nr_of_customers
avg_sales_per_customer
```

```python
# How many orders monthly?
# I will use another new table to get the date information.

query_orders_monthly = "SELECT order_id, order_purchase_timestamp FROM olist.orders"
orders_monthly = pd.read_sql_query(query_orders_monthly, db_connection)
orders_monthly
```

```python
# Checking the missing values.

order_year_price_join.isna().sum()
```

```python
# We need to break the datetime to get months. We will add a new column to the table which only has the month inside.
```

```python
orders_monthly["month"] = orders_monthly["order_purchase_timestamp"].dt.month
```

```python
orders_monthly
```

```python
# Top categories by sales amount. Here I will join two tables to get necessary information.

query_top_categories = "SELECT * FROM olist.order_payments Inner join olist.order_items on order_payments.order_id = order_items.order_id"
top_categories_first = pd.read_sql_query(query_top_categories, db_connection)
top_categories_first
```

```python
# This join is also for finding the top categories by sales.

query_top_categories_new = "SELECT * FROM olist.products Inner join olist.product_category_name_translation on products.product_category_name = product_category_name_translation.product_category_name"
top_categories_second = pd.read_sql_query(query_top_categories_new, db_connection)
top_categories_second
```

```python
# Here I will merge the two previous ones together.

top_categories_df = pd.merge(top_categories_first, top_categories_second, on = "product_id")
top_categories_df
```

```python
# Dropping unnecessary columns.

top_categories_df.drop(columns = ["payment_sequential", "payment_type", "payment_installments", "order_item_id", "shipping_limit_date", "product_name_length", "product_description_length", "product_photos_qty", "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"], inplace = True)

top_categories_df
```

```python
# Checking for missing values.

top_categories_df.isna().sum()
```

```python
# A bit of descriptive statistics about the table.

top_categories_df.describe()
```

```python
# Here is the top 10 categories by sales amount.
# Grouping by category using the sum of payment value. 
# Making two new columns with total value and the percentage.

category_sold = top_categories_df.groupby(["product_category_name_english"], as_index=False)["payment_value"].count().rename(columns = {"payment_value":"total_value"})
category_sold["%"] = round((category_sold["total_value"] / category_sold["total_value"].sum()) * 100, 2)
category_sold10 = category_sold.nlargest(10, "total_value")
category_sold10
```

```python
# Joining the tables needed for top 10 sellers.

query_for_top_sellers = "SELECT * FROM olist.order_items inner join olist.order_payments on order_items.order_id = order_payments.order_id"
top_sellers_first = pd.read_sql_query(query_for_top_sellers, db_connection)
top_sellers_first
```

```python
# Here is another column needed for the top 10 sellers.

query_for_top_sellers_two = "SELECT seller_id FROM olist.sellers"
top_sellers_second = pd.read_sql_query(query_for_top_sellers_two, db_connection)
top_sellers_second
```

```python
# Merging them together.

top_sellers_df = pd.merge(top_sellers_first, top_sellers_second, on = "seller_id")
top_sellers_df
```

```python
# Dropping unnecessary columns.

top_sellers_df.drop(columns = ["order_item_id", "shipping_limit_date", "payment_sequential", "payment_type", "payment_installments"], inplace = True)

top_sellers_df
```

```python
# Checking missing values.

top_sellers_df.isna().sum()
```

```python
# Small descriptive table.

top_sellers_df.describe()
```

#### Here I will start joining, exploring and cleaning all tables which I need for for the second dashboard.

```python
# I will make a table for delivery times, customers and their location.
```

```python
# First, geolocation information.

query_geolocation = "SELECT * FROM olist.geolocation"
geolocation_df = pd.read_sql_query(query_geolocation, db_connection)
geolocation_df
```

```python
# Customer information.

query_order_delivery = "SELECT * FROM olist.orders inner join olist.customers on orders.customer_id=customers.customer_id"
order_delivery_df = pd.read_sql_query(query_order_delivery, db_connection)
order_delivery_df
```

```python
# Renaming the customer zip code column so it will match with the geolocation table.

order_delivery_df = order_delivery_df.rename(columns={"customer_zip_code_prefix":"geolocation_zip_code_prefix"})
order_delivery_df
```

```python
# Merging two previous tables together.

location_delivery_df = pd.merge(order_delivery_df, geolocation_df, on = "geolocation_zip_code_prefix", how = "right")
location_delivery_df
```

```python
# The previously made table needs cleaning so it will have the correct information.
# I will delete all duplicates for unique customer id-s. It will show only one purchase per customer.

location_delivery_unique = location_delivery_df.drop_duplicates("customer_unique_id").reset_index(drop=True)
location_delivery_unique
```

```python
# I will make a new column to know if the order was delayed or not, using the date-time columns.

location_delivery_unique["if_delayed"] = location_delivery_unique["order_delivered_customer_date"] > location_delivery_unique["order_estimated_delivery_date"]
```

```python
# Checking the outcome.

location_delivery_unique
```

```python
# Checking missing values.

location_delivery_unique.isna().sum()
```

```python
# Dropping all rows with missing values.

location_delivery_unique.dropna(axis = 0, inplace = True)
```

```python
# Checking the outcome.

location_delivery_unique.isna().sum()
```

```python
# Here I will clean the location and delivery tabel again based on customer id (not unique id like before).
# Here I will delete all duplicates according to customer id, which are all not unique, but show all purchases.

location_delivery_customer = location_delivery_df.drop_duplicates("customer_id").reset_index(drop=True)
location_delivery_customer
```

```python
# Adding a new column for if the order was delayed.

location_delivery_customer["is_delayed"] = location_delivery_customer["order_delivered_customer_date"] > location_delivery_customer["order_estimated_delivery_date"]
```

```python
# Checking the outcome.

location_delivery_customer
```

```python
# Checking for missing values.

location_delivery_customer.isna().sum()
```

```python
# Dropping rows with missing values.

location_delivery_customer.dropna(axis = 0, inplace = True)
```

```python
# Checking the outcome.

location_delivery_customer.isna().sum()
```

```python
# Getting the necessary freight information.

fright_delivery = "SELECT * FROM olist.order_items"
fright_df = pd.read_sql_query(fright_delivery, db_connection)
fright_df
```

```python
# In order to compare the fright values to markets, I will join relevant tables.

location_fright_df = pd.merge(location_delivery_customer, fright_df, on = "order_id")
location_fright_df
```

```python
# Saving the table as csv files.

# location_fright_df.to_csv(r'Data\location and fright df.csv', index = False)
```

```python
# Calculating the delivery days for each order. Adding the result to a new column.

location_fright_df["delivery_days"] = (location_fright_df["order_delivered_customer_date"] - location_fright_df["order_purchase_timestamp"]).dt.days
                                                                                                          
```

```python
# Checking the result.

location_fright_df
```

```python
# Checking for missing values.

location_fright_df.isna().sum()
```
